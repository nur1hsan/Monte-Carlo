```python
import numpy as np                        # Import library numpy

m = 0                                     # Inisiasi nilai m (total titik yang juga diartikan sebagai jumlah iterasi)
n = 0                                     # Inisiasi nilai n (titik yang berada di dalam lingkaran berjari-jari 1)

x_in = []                                 # Inisiasi array titik x yang berada di dalam lingkaran
y_in = []                                 # Inisiasi array titik y yang berada di dalam lingkaran
x_out = []                                # Inisiasi array titik x yang berada di luar lingkaran
y_out = []                                # Inisiasi array titik y yang berada di luar lingkaran
pi_arr = []                               # Inisiasi array nilai pi hasil perhitungan
m_arr = []                                # Inisiasi array jumlah iterasi

for i in range (100000):                  # Looping untuk titik sejumlah 100000 titik
  x = np.random.uniform(low=-1, high=1)   # x sebagai random number, dengan batas -1 sampai 1
  y = np.random.uniform(low=-1, high=1)   # y sebagai random number, dengan batas -1 sampai 1
  m = m + 1                               # Menghitung jumlah iterasi yang dilakukan
  m_arr.append(m)                         

  if (x**2 + y**2 < 1):                   # Syarat apabila titik berada di dalam lingkaran berjari-jari 1
    n = n + 1                             # Menghitung secara berkala jumlah titik yang berada di dalam lingkaran
    x_in.append(x)                        # Memasukkan nilai x yang berada di dalam lingkaran dalam array x_in
    y_in.append(y)                        # Memasukkan nilai y yang berada di dalam lingkaran dalam array y_in

  else: #(x**2 + y**2 > 1)                # Syarat apabila titik berada di luar lingkaran berjari-jari 1
    x_out.append(x)                       # Memasukkan nilai x yang berada di luar lingkaran dalam array x_out
    y_out.append(y)                       # Memasukkan nilai y yang berada di luar lingkaran dalam array y_out

  pi = 4 * (n/m)                          # Menghtiung pi
  pi_arr.append(pi)
  print(pi)                               # Menuliskan nilai pi yang diperoleh

  if (pi == 3.1415926535):                                                  # Syarat ketika nilai pi sesuai dengan nilai referensi (diambil dari www.math.com)
    print("Nilai pi referensi didapat pada iterasi ke", m)
  elif (m == 100000) and (pi != 3.1415926535):                              # Syarat apabila telah dilakukan iterasi terakhir, namun nilai pi yang didapat belum  sesuai dengan nilai referensi
    print("Sampai iterasi ke", m,"nilai pi referensi belum diperoleh")
```

```python
# Memvisualisasikan hubungan jumlah iterasi (dalam logaritmik) terhadap nilai pi referensi

import matplotlib.pyplot as plt                         # Import library matplotlib.

plt.figure(figsize=[7,7])
plt.plot(np.log10(m_arr), pi_arr, c="midnightblue", label="Pendekatan Nilai Pi")
plt.axhline(y = 3.1415926535, c="maroon", label="Pi referensi (3.1415926535)")
plt.xlabel("m dalam logaritmik")
plt.ylabel("Pi perhitungan")
plt.legend(loc='lower right')
plt.show
```

**Terlihat pada grafik, semakin besar m, nilai pi semakin mendekati nilai referensi**

```python
# Menghitung Galat

pi_ref = 3.1415926535                                 # Nilai pi referensi (dari www.math.com)

Galat = pi_ref - pi                                   # Menghitung Galat
Galat_persen = abs((Galat/pi_ref) * 100)                 # Menghitung Galat Relatif

print("True Error =", Galat)
print("Relative True Error =", Galat_persen, "%")
```

```python
# Memvisualisasikan nilai n dan m

plt.figure(figsize=[7,7])                               # Menentukan ukuran grafik
plt.scatter(x_in, y_in, s=3, c="lightpink")             # Mem-plot titik yang berada di dalam lingkaran
plt.scatter(x_out, y_out, s=3, c="midnightblue")        # Mem-plot titik yang berada di luar lingkaran
plt.xlabel("x")
plt.ylabel("y")
plt.show                                                # Menampilkan grafik
```

**Pada grafik, n menggambarkan jumlah titik yang berwarna merah muda, sementara m menggambarkan jumlah titik yang berwarna merah muda dan biru tua (keseluruhan titik).**
